```python
S = []
top = None 


def is_empty(stk):
    if stk == [] :
        return True
    else:
        return False 
    
def push(stk,value):
    stk.append(value)    
    top = len(stk)-1

def pop(stk):
    if is_empty(stk):
        return ("underflow")
    else:
        i = stk.pop()
        if len(stk) == 0:
            top = None
        else:
            top -= 1
        return i 

def peak(stk):
    
    if len(stk)== 0 :
        return ("underflow")
    else:
        top = len(stk)-1
        return stk[top]


def display(stk):
    if is_empty(stk):
        print("stack is empty")
    else:
        top  = len(stk)-1
        print(stk[top], "<---top")
        for i in range(top-1,-1,-1):
            print(stk[i])

```

```python
while True :
    print("\n \n")
    print("Stack implementation")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Display")
    print("5. Exit")
    print("\n \n")

    ch = int(input("Enter the choice(1-5) : "))



    if ch==1:
        value = int(input("Enter the value: "))
        push(S,value)
        print(f'{value} added successfully.')
        input("press any key to continue....")


    if ch==2:
        value = pop(S)
        if value == "underflow":
            print("stack is empty")
        else:
            print(f"{value} is popped")
        input("press any key to continue....")

    if ch == 3 :
        value = peak(S)
        if value == "underflow":
            print("stack is empty")
        else:
            print(f"{value} is at the top")
        input("press any key to continue....")

    if ch==4 :
        display(S)
        
        input("press any key to continue....")

    elif ch==5: break

    else:print("choose from 1-5")
```
